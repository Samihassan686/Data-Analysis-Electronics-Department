## Importing Necessary Libraries

```python
import pandas as pd
```

## Loading Data

```python
df = pd.read_json('Data\electronics.json')
null_summary = df.head()
null_summary
```

## Analyzing Dataset
Our dataset contains emtpy strings in form of "", so we replace them with NA values using df.replace function. Then we count the total null values per column and store that count is a dataframe.

```python
df.replace("", pd.NA, inplace=True)
null_info = df.isnull().sum()
null_summary = pd.DataFrame({'NA Values': null_info})
null_summary
```
